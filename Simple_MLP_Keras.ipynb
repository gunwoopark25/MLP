```python
###제일 중요한 과정###
#Data 준비 (내가 직접 해야하는 과정)

from tensorflow.keras import utils

(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

X_train = X_train.reshape(60000, 28*28) / 255.0
X_test = X_test.reshape(10000, 784) / 255.0
Y_train = utils.to_categorical(Y_train)
Y_test = utils.to_categorical(Y_test)
```

```python
#라이브러리를 활용해서 model 형성

from keras import models
from keras import layers

model = models.Sequential() #모델의 구조 구성 input-> layer1 -> layer2 -> output
model.add(layers.Dense(200, activation="relu", input_shape=(28*28,))) #Dense레이어에 node의 개수 200개, relu를 사용한다. input data의 shape이 28*28 형태이다.
model.add(layers.Dense(32, activation="relu"))
model.add(layers.Dense(10, activation="softmax"))

#학습시키는 과정
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

model.fit(X_train, Y_train, epochs=10, batch_size=32)

#test에 대해서 확인
loss_and_metrics = model.evaluate(X_test, Y_test, batch_size=32)

classes = model.predict(X_test, batch_size=32)
```
